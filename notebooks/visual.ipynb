```python
FILE_ID = "1NaXgQ0LiW-RjNohEZ_uwRw19E-czPizL"  # ID файла на Google Drive
file_url = f"https://drive.google.com/uc?id={FILE_ID}"

df = pd.read_csv(file_url)  # читаем файл

df.head(5)  # выводим на экран первые 5 строк для проверки
```

```python
sns.histplot(data=df, x="Year First Published")
```

По этому графику видно, что пик открытий приходится на 21 век.

```python
plt.figure(figsize=(12, 6))
top_countries = df['Country of Type Locality'].value_counts()
top_countries = top_countries.head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Топ-10 стран по количеству минералов')
plt.xlabel('Количество минералов')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
status_counts = df['IMA Status'].value_counts()
sns.barplot(x=status_counts.values, y=status_counts.index)
plt.title('Распределение по статусу IMA')
plt.xlabel('Количество')
plt.show()
```

```python
from collections import Counter

all_elements = []
for elements in df['Chemistry Elements'].dropna():
    all_elements.extend(elements.split())

element_counts = Counter(all_elements)
top_elements = dict(element_counts.most_common(20))

plt.figure(figsize=(12, 8))
sns.barplot(x=list(top_elements.values()), y=list(top_elements.keys()))
plt.title('Топ-20 самых частых химических элементов в минералах')
plt.xlabel('Количество минералов')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
year_status = df.groupby(['Year First Published', 'IMA Status']).size().reset_index(name='count')
sns.lineplot(data=year_status, x='Year First Published', y='count', hue='IMA Status')
plt.title('Динамика открытия минералов по годам и статусу')
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_data = df.pivot_table(
    index='Crystal Systems', 
    columns='IMA Status', 
    values='Oldest Known Age (Ma)', 
    aggfunc='count',
    fill_value=0
)

plt.figure(figsize=(24, 16))
sns.heatmap(pivot_data, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Распределение минералов по кристаллическим системам и статусу IMA')
plt.tight_layout()
plt.show()
```
